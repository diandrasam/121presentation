```python
import cmath

def function(x):
    f = -x**4 + 2*(x**3) - x**2 - 2*x + 2
    return f
```

```python
   def muller(point0, point1, point2, tolerance, max_iterations):
    h1 = point1 - point0
    h2 = point2 - point1
    d1 = (function(point1) - function(point0)) / h1
    d2 = (function(point2) - function(point1)) / h2
    
    a = (d2 - d1) / (h2 + h1)
    c = function(point2)
    i = 3
    
    while(i <= max_iterations):
        b = d2 + h2*a
        D = cmath.sqrt((b**2) - 4*a*c)

        if abs(b-D) < abs(b+D):
            E = b+D
        else:
            E = b-D
        h = (-2* c) / E
        p = point2 + h

        if abs(h) < tolerance:
            print(i)
            print(p)
            break
        else:
            print(i)
            print(p)
            point0 = point1
            point1 = point2
            point2 = p
            h1 = point1 - point0
            h2 = point2 - point1
            d1 = (function(point1) - function(point0)) / h1
            d2 = (function(point2) - function(point1)) / h2

            a = (d2 - d1) / (h2 + h1)
            c = function(point2)
            i += 1
    if(i > max_iterations):
        print("Method unsuccessful")
```
